```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulka's Yearly Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparision")
plt.legend()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
rohit = [0, 200, 200, 600, 1100, 1200, 1300, 1800, 2200, 2600]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, linestyle='--', label="Virat Kohli")
plt.plot(years, rohit, linestyle='--', label="Rohit sharma")
plt.plot(years, sehwag, linestyle='--',  label= "Virender Sehwag")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Performance Comparision")
plt.legend()
```
