```python
import matplotlib.pyplot as plt
 
plt.style.use("ggplot")
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
runs = [18000, 8000, 12000, 9500]
explode = [0.1, 0, 0, 0] 
```

```python
plt.pie(runs, 
        labels=labels,
        colors=colors,
        explode=explode,
        wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'})
```

```python
plt.pie(runs, 
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True,
        startangle=100,
        autopct="%1.0f%%"
        )
```
